```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Function: y = (x + 3)^2
def f(x):
    return (x + 3) ** 2

# Derivative (gradient): dy/dx = 2(x + 3)
def gradient(x):
    return 2 * (x + 3)
```

# Gradient Descent Implementation

```python
# Gradient Descent Implementation

def gradient_descent(start_x, learning_rate=0.1, max_iters=50, tol=1e-6):
    x = start_x
    steps = [x]   # store all iteration values

    for i in range(max_iters):
        grad = gradient(x)
        x_new = x - learning_rate * grad
        steps.append(x_new)

        print(f"Iteration {i+1}: x = {x_new}, f(x) = {f(x_new)}")

        if abs(x_new - x) < tol:  # stop condition
            break

        x = x_new

    return x, f(x), steps
```

```python
start_x = 2  # starting point as per the problem statement

min_x, min_y, steps = gradient_descent(start_x)

print("\n=======================")
print(f"Final Minima at x = {min_x}")
print(f"Minimum y = {min_y}")
print("=======================\n")
```

```python
# Plot the quadratic curve and the descent movement

x_range = np.linspace(-6, 2, 100)   # range of x values
plt.plot(x_range, f(x_range), label="y = (x + 3)^2")
plt.scatter(steps, [f(x) for x in steps])
plt.title("Gradient Descent â€” Finding Local Minima")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```
